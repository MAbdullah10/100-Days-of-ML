```python
df_list = []

for i in range(1, 429):
    response = requests.get(
        "https://api.themoviedb.org/3/movie/top_rated?api_key=58f52024afa11762e09d0fac215dcece&language=en-US&page={}".format(i)
    )

    temp_df = pd.DataFrame(response.json()['results'])[
        ['id', 'title', 'release_date', 'popularity', 'vote_average']
    ]

    df_list.append(temp_df)

df = pd.concat(df_list, ignore_index=True)

df.head()
```
