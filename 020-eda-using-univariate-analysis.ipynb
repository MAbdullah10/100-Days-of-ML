```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd 
df = pd.read_csv('./src/kashti.csv')
df.head()
```

# 01 Categorical Data

### a. Countplot

```python
sns.countplot(df['sex'])
df['sex'].value_counts()
```

### b. Piechart

```python
df['who'].value_counts().plot(kind = 'pie',autopct = "%.2f")
```

# 02 Numerical Data

### a. Histogram

```python
plt.hist(df['age'],bins=10)
```

### b. Distplot

```python
sns.distplot(df['age'])
```

### c. Boxplot

```python
sns.boxplot(df['age'])
```

```python
df['age'].skew()
```
