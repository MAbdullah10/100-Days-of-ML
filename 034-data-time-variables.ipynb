```python
import numpy as np
import pandas as pd
```

```python
date = pd.read_csv("./src/orders.csv")
date
```

```python
date.info()
```

```python
date['date'] =pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
#extracting the year, month and day from the date column
date['year'] = date['date'].dt.year
date['month'] = date['date'].dt.month
date['day'] = date['date'].dt.day
date.head()
```

```python
# extract month name from the date column
date['month_name'] = date['date'].dt.month_name()
date.head()
```

```python
# day of the week - name
date['day_of_week'] = date['date'].dt.day_name()
date.head()
```

```python
#day of the week - number
date['day_of_week_num'] = date['date'].dt.dayofweek # Monday = 0, Sunday = 6
date.head()
```

```python
# is inweekend?
date['is_weekend'] = date['day_of_week_num'].apply(lambda x: 1 if x >= 5 else 0)
date.head()
```

```python
# week of the year
date['week_of_year'] = date['date'].dt.isocalendar().week
date.head()
```

```python
# extract quarter of the year
date['quarter'] = date['date'].dt.quarter
date.head()
```

```python
# extract semester of the year
date['semester'] = date['date'].dt.month.apply(lambda x: 1 if x <= 6 else 2) 
date.head()
```

```python
#extract time elapsed between dates
import datetime
today = datetime.datetime.today()
today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# months passed
(today - date['date']).dt.days / 30
```
