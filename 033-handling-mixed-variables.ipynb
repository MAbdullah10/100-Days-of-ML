```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./src/mixed_varaibles.csv")
df.head()
```

```python
fig = df['number'].value_counts().plot.bar()
```

```python
df['numerical_number'] = pd.to_numeric(df['number'], errors='coerce',downcast='integer')
df['categorical_number'] = np.where(df['numerical_number'].isnull(), df['number'], np.nan)
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Cabin_num'] = df['Cabin'].str.extract('(\d+)')
df['Cabin_cat'] = df['Cabin'].str[0]
df.head()
```

```python
df['Cabin_cat'].value_counts().plot.bar()
```
