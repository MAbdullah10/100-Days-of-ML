```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
requests.get("https://www.ambitionbox.com/list-of-companies?pages=1")
```

```python
requests.get("https://www.ambitionbox.com/list-of-companies?pages=1").text
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get("https://companiesmarketcap.com/pakistan/largest-companies-in-pakistan-by-market-cap/", headers = headers).text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
soup.find_all('h1')[0].text
```

```python
soup.find_all('th')
```

```python
soup.find_all('td')
```

```python
table = soup.find("table")
rows = table.find_all("tr")

# Extract data
data = []
for row in rows[1:]:  # skip header
    cols = row.find_all("td")
    if len(cols) >= 7:  # ensure enough columns
        rank = cols[0].text.strip()
        name = cols[1].text.strip()
        market_cap = cols[2].text.strip()
        price = cols[3].text.strip()
        country = cols[6].text.strip()
        data.append([rank, name, market_cap, price, country])

# Create DataFrame
df = pd.DataFrame(data, columns=["Rank", "Name", "Market Cap", "Price", "Country"])

# Show top rows
df.head()
```

```python
data = []
for page in range(1, 4):  # first 3 pages
    url = f"https://companiesmarketcap.com/pakistan/largest-companies-in-pakistan-by-market-cap/?page={page}"
    response = requests.get(url, headers={"User-Agent":"Mozilla/5.0"})
    soup = BeautifulSoup(response.text, "lxml")
    table = soup.find("table")
    for row in table.find_all("tr")[1:]:
        cols = row.find_all("td")
        if len(cols) >= 7:
            data.append([cols[0].text.strip(), cols[1].text.strip(), cols[2].text.strip()])

data
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage2 = requests.get("https://books.toscrape.com", headers = headers).text
```

```python
soup2=BeautifulSoup(webpage2,'lxml')
soup2
```

```python
soup2.find_all('h3')
```

```python
products = soup2.find_all("article", class_="product_pod")

data = []

for p in products:
    title = p.h3.a["title"]
    price = p.find("p", class_="price_color").text
    
    data.append([title, price])

df = pd.DataFrame(data, columns=["Title", "Price"])

df.head(10)
```

```python
df.to_csv("./output/books.csv", index=False)
```
