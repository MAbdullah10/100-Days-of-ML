```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./src/Social_Network_Ads.csv")
df
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop('Purchased',axis=1),df['Purchased'],test_size=0.3,random_state=0)

X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()
scalar.fit(X_train)

X_train_Scaled = scalar.transform(X_train)
X_test_scaled = scalar.transform(X_test)

scalar.mean_
```

```python
X_train
```

```python
X_train_Scaled
```

- has converted into numpy array while standardization so need to convert it back into dataframe

```python
X_train_Scaled = pd.DataFrame(X_train_Scaled,columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

```python
X_test_scaled
```

```python
np.round(X_train.describe(),1)
```

```python
np.round(X_train_Scaled.describe(),1)
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(12,5))

ax1.set_title('Before Standard Scaling')
ax1.scatter(X_train['Age'],X_train['EstimatedSalary'],color='red')

ax2.set_title('After Standard Scaling')
ax2.scatter(X_train_Scaled['Age'],X_train_Scaled['EstimatedSalary'],color='green')
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(12,5))

ax1.set_title('Before Standard Scaling')
sns.kdeplot(X_train['Age'],ax=ax1)
sns.kdeplot(X_train['EstimatedSalary'],ax=ax1)

ax2.set_title('After Standard Scaling')
sns.kdeplot(X_train_Scaled['Age'],ax=ax2)
sns.kdeplot(X_train_Scaled['EstimatedSalary'],ax=ax2)
plt.show()
```

## Why scaling is important

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
lr_scaled.fit(X_train_Scaled,y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred_scaled = lr_scaled.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Actual ",accuracy_score(y_test,y_pred))
print("Scaled ",accuracy_score(y_test,y_pred_scaled))
```
